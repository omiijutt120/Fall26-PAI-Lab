## **Task - Text Preprocessing**

Text preprocessing cleans raw text before any NLP model uses it. Steps covered: tokenization, stopword removal, stemming, and lemmatization.

## **Imports**

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

## **Tokenization**

Splits a sentence into individual words.

```python
text = "The dog barked loudly at the stranger outside the gate."

tokens = word_tokenize(text)
print("Tokens:", tokens)
```

## **Stopword Removal**

Removes common words like 'the', 'at', 'is' that carry no real meaning.

```python
stop_words = set(stopwords.words('english'))

filtered = [w for w in tokens if w.lower() not in stop_words]
print("After removing stopwords:", filtered)
```

## **Stemming**

Reduces words to their root form. The result may not always be a real word.

```python
stemmer = PorterStemmer()

stemmed = [stemmer.stem(w) for w in filtered]
print("After stemming:", stemmed)
```

## **Lemmatization**

Also reduces words to their base form, but always returns a valid word using grammar rules.

```python
lemmatizer = WordNetLemmatizer()

lemmatized = [lemmatizer.lemmatize(w, pos='v') for w in filtered]
print("After lemmatization:", lemmatized)
```

## **Full Pipeline**

All four steps applied to a new sentence in one go.

```python
def preprocess(text):
    tokens = word_tokenize(text)
    tokens = [w for w in tokens if w.isalpha()]
    tokens = [w for w in tokens if w.lower() not in stop_words]
    stemmed = [stemmer.stem(w) for w in tokens]
    lemmatized = [lemmatizer.lemmatize(w, pos='v') for w in tokens]
    return tokens, stemmed, lemmatized

sentence = "The children were running and playing games in the park."
clean, stemmed, lemmatized = preprocess(sentence)

print("Cleaned tokens :", clean)
print("Stemmed        :", stemmed)
print("Lemmatized     :", lemmatized)
```
